## Find the min , max , mean and mode on the data

```python
print("Min = ",aData['age'].min())
print("Max = ",aData['age'].max())
print("Mean = ",aData['age'].mean())
print("Mode = ",aData['age'].mode()[0])
```

```python
print("Min = ",aData['hours-per-week'].min())
print("Max = ",aData['hours-per-week'].max())
print("Mean = ",aData['hours-per-week'].mean())
print("Mode = ",aData['hours-per-week'].mode()[0])
```

```python
plt.scatter(aData['age'],aData['hours-per-week'])
plt.xlabel('age')
plt.ylabel('hours-per-week')
plt.show()
```

```python
plt.scatter(aData['age'],aData['workclass'])
plt.xlabel('age')
plt.ylabel('work class')
plt.show()
```

### Analyze the person's work on the week

```python
aAvgAgeByHPW = aData.groupby('age')['hours-per-week'].mean()
aAvgAgeByHPW.head()
```

```python
print(aAvgAgeByHPW.values.max())
print(aAvgAgeByHPW.values.min())
```

```python
plt.plot(aAvgAgeByHPW.index,aAvgAgeByHPW.values)
plt.xlabel('age')
plt.ylabel('hours-per-week')
plt.grid(True,linestyle='--', alpha=0.5)
plt.show()
```

```python
plt.scatter(aAvgAgeByHPW.index,aAvgAgeByHPW.values)
plt.xlabel('age')
plt.ylabel('hours-per-week')

plt.show()
```

### Analyze the person's with there age WorkClass

```python
aDataAvgAgeByWORKCLASS = aData.groupby('age')['workclass'].apply(lambda x: x.mode()[0])
aDataAvgAgeByWORKCLASS
```

```python
plt.plot(aDataAvgAgeByWORKCLASS.index,aDataAvgAgeByWORKCLASS.values)
plt.xlabel('age')
plt.ylabel('workclass')
plt.show()
```

```python
plt.scatter(aDataAvgAgeByWORKCLASS.index,aDataAvgAgeByWORKCLASS.values)
plt.xlabel('age')
plt.ylabel('workclass')
plt.show()
```

```python
aData['income'].unique()
```

```python
aData['ModifyIncome'] = aData['income'].apply(lambda x: 1 if x == '>50K' else 0)
aData['ModifyIncome'].unique()
```

```python
aDataIncomeWithAge = aData.groupby('age')['ModifyIncome'].value_counts()
aDataIncomeWithAge
```

```python
aDataIncomeWithAge.index.get_level_values(0)
```

```python
aDataIncomeWithAge.index.get_level_values(1)
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(aDataIncomeWithAge.index.get_level_values(0),aDataIncomeWithAge.values,c=aDataIncomeWithAge.index.get_level_values(1),alpha=0.9)
plt.title('Income Distribution by Age')
plt.xlabel('age')
plt.ylabel('Count')
plt.grid(True,linestyle='--', alpha=0.5)
crb = plt.colorbar()
crb.set_label('Income Category (0: <=50K, 1: >50K)')
plt.show()
```
