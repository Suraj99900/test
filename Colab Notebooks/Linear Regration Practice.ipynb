```python
import pandas as pd
import numpy as np
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from matplotlib import pyplot as plt
```

```python
df = fetch_california_housing(as_frame=True)
```

```python
df
```

```python
df_data = df.data
```

```python
df_data
```

```python
df_data.columns = df.feature_names
```

```python
df_data.head()
```

```python
x = df_data
y = df.target
```

```python
print(x.shape)
print(y.shape)
```

```python
print(x.info())
print(y.info())
```

```python
print(x.isnull().sum())
print(y.isnull().sum())
```

```python
# Spliting The train and test data
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=42)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# Standar scaler process
from sklearn.preprocessing import StandardScaler
```

```python
scalerTrain = StandardScaler()
scalerTest = StandardScaler()

x_train = scalerTrain.fit_transform(x_train)
x_test = scalerTest.fit_transform(x_test)
```

```python
x_train
```

```python
# #### corss validation
from sklearn.model_selection import cross_val_score
```

```python
regression = LinearRegression()
```

```python
regression.fit(x_train, y_train)
```

```python
mse = cross_val_score(regression, x_train, y_train, cv=10, scoring='neg_mean_squared_error')
```

```python
np.mean(mse)
```

```python
### predection
y_pred = regression.predict(x_test)
```

```python
y_pred
```

```python
y_test
```

```python
import seaborn as sns
```

```python
sns.displot(y_test - y_pred,kind='kde')
plt.show()
```

```python
from sklearn.metrics import r2_score
```

```python
print("r2_score :- ",r2_score(y_test, y_pred))
```
