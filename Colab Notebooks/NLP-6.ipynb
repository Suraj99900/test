```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import speech_recognition as sr
from pydub import AudioSegment
```

```python
def speech_to_text(audio_file):
    """Converts speech in an audio file to text."""
    recognizer = sr.Recognizer()
    with sr.AudioFile(audio_file) as source:
        audio = recognizer.record(source)
    try:
        text = recognizer.recognize_google(audio)
        return text
    except sr.UnknownValueError:
        return "Could not understand audio"
    except sr.RequestError as e:
        return f"Could not request results from Google Speech Recognition service; {e}"
```

```python
def answer_matching(student_answer, ideal_answer):
    """Compares student's answer with the ideal answer."""
    # Basic implementation: Check if student's answer contains keywords from the ideal answer
    ideal_keywords = ideal_answer.lower().split()
    student_keywords = student_answer.lower().split()

    # Calculate the percentage of ideal keywords present in the student's answer
    match_percentage = len(set(ideal_keywords) & set(student_keywords)) / len(ideal_keywords) * 100

    # Define a threshold for a correct answer (e.g., 70%)
    threshold = 70

    if match_percentage >= threshold:
        return True  # Answer is considered correct
    else:
        return False  # Answer is considered incorrect
```
