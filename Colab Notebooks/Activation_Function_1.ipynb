```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10,10,100)
```

```python
def linear(x):
    return x  ## Outputs the input directly without any transformation.
```

```python
plt.plot(x,linear(x))
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid()
plt.show()

### this draw a linear graph 
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
# Lets draw a graph for sigmoid

plt.figure(figsize=(5,5))
plt.plot(x,sigmoid(x))
plt.title("Sigmoid")
plt.xlabel("x-Input")
plt.ylabel("y-Output")
plt.grid()
plt.show()
```

```python
def tanh(x):
    return (np.exp(x)-np.exp(-x))/(np.exp(x)+np.exp(-x))
```

```python
plt.figure(figsize=(5,5))
plt.plot(x,tanh(x))
plt.title("Tanh")
plt.xlabel("Input")
plt.ylabel("Output")
plt.grid()
plt.show()
```

```python
# Relu (Rectified Lineear unit )
def relu(x):
    return np.maximum(0,x)
```

```python
plt.figure(figsize = (5,5))
plt.plot(x,relu(x))
plt.title("Tanh")
plt.xlabel("Input")
plt.ylabel("Output")
plt.grid()
plt.show()
```

```python
# Softmax
def softmax(x):
    e_x = np.exp(x-np.max(x))  ## # subtract max for numerical stability
    return (e_x/e_x.sum())
```

```python
plt.figure(figsize = (5,5))
plt.plot(x,softmax(x))
plt.title("Softmax")
plt.xlabel("Input")
plt.ylabel("Output")
plt.grid()
plt.show()
```

```python
def leakyRelu(x,alpha=0.1):
     return np.where(x > 0, x, alpha * x)
```

```python
plt.figure(figsize = (5,5))
plt.plot(x,leakyRelu(x,0.03))
plt.title("Leaky Relu")
plt.xlabel("Input")
plt.ylabel("Output")
plt.grid()
plt.show()
```
