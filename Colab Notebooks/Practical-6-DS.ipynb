```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score

```

```python
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['target'] = iris.target
iris_df.head()
```

```python
iris_df.shape
```

```python
iris_df.info()
```

```python
iris_df.isnull().sum()
```

```python
x = iris_df[['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)']]
y = iris_df['target']
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
# Train Naïve Bayes classifier
nb_model = GaussianNB()
nb_model.fit(X_train, y_train)
```

```python
print("Train Score : ",nb_model.score(X_train, y_train)*100,"%")
```

```python
y_pred = nb_model.predict(X_test)
```

```python
print("y_pred Data :- ",y_pred)
```

```python
print("y_test Data :- ",y_test.values)
```

```python
print("Test Score : ",nb_model.score(X_test, y_test)*100,"%")
```

```python
print("Cpnfusion Matrix :- ")
print(confusion_matrix(y_test, y_pred))
confusion_matrix = confusion_matrix(y_test, y_pred)
```

```python
print("Accuracy : ",accuracy_score(y_test, y_pred)*100,"%")
```

```python
print("Precision : ",precision_score(y_test, y_pred, average='macro')*100,"%")
```

```python
print("Recall : ",recall_score(y_test, y_pred, average='macro')*100,"%")
```

```python
print("Error Rate : ",(1-accuracy_score(y_test, y_pred))*100,"%")
```
