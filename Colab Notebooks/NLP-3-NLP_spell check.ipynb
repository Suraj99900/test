```python
import difflib

def min_edit_distance(str1, str2):
    m = len(str1)
    n = len(str2)
    # Initialize a matrix to store distances
    dp = [[0 for _ in range(n + 1)] for _ in range(m + 1)]


    # Fill base cases
    for i in range(m + 1):
        dp[i][0] = i  # Deletion cost
    for j in range(n + 1):
        dp[0][j] = j  # Insertion cost

    # Compute minimum edit distance
    for i in range(1, m + 1):
        for j in range(1, n + 1):
            if str1[i - 1] == str2[j - 1]:
                dp[i][j] = dp[i - 1][j - 1]  # No change needed
            else:
                dp[i][j] = 1 + min(dp[i - 1][j],    # Deletion
                                   dp[i][j - 1],    # Insertion
                                   dp[i - 1][j - 1] # Substitution
                                  )

    return dp[m][n]

def spell_correction(word, dictionary):
    closest_match = min(dictionary, key=lambda w: min_edit_distance(word, w))
    return closest_match

# Example usage
if __name__ == "__main__":
    dictionary = ["hello", "world", "python", "programming", "edit", "distance"]
    word = input("Enter a misspelled word: ")
    corrected_word = spell_correction(word, dictionary)
    print(f"Did you mean: {corrected_word}?")
```
