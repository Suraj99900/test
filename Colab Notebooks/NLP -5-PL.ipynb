```python
import speech_recognition as sr
from gtts import gTTS
import os
import random
```

```python
# Sample recommendation data
recommendations = {
    "movies": ["Inception", "Interstellar", "The Matrix", "Blade Runner 2049"],
    "books": ["Dune", "1984", "Brave New World", "The Hitchhiker's Guide to the Galaxy"],
    "music": ["Bohemian Rhapsody", "Stairway to Heaven", "Imagine", "Hotel California"],
    "hi":["how are you..."]
}
```

```python
def recognize_speech():
    """Listen to the user's voice and convert it to text using Google Speech API."""
    recognizer = sr.Recognizer()
    with sr.Microphone() as source:
        print("ðŸŽ¤ Speak now...")
        recognizer.adjust_for_ambient_noise(source)
        try:
            # Listen with a 10-second timeout
            audio = recognizer.listen(source, timeout=10)
        except sr.WaitTimeoutError:
            print("Listening timed out. No speech detected.")
            return "Sorry, I could not hear anything."

    try:
        # Use Google Web Speech API
        text = recognizer.recognize_google(audio)
        print("ðŸ—£ You said:", text)
        return text.lower()
    except sr.UnknownValueError:
        return "Sorry, I could not understand the audio."
    except sr.RequestError:
        return "Could not request results, please check your connection."

def text_to_speech(text):
    """Convert chatbot response to speech."""
    tts = gTTS(text=text, lang='en')
    tts.save("response.mp3")
    os.system("start response.mp3" if os.name == "nt" else "mpg321 response.mp3")

def recommend(user_input):
    """Provide recommendations based on user input."""
    if "movie" in user_input:
        category = "movies"
    elif "book" in user_input:
        category = "books"
    elif "music" in user_input:
        category = "music"
    elif "hi" in user_input:
        category = "hi"
    else:
        return "I can recommend movies, books, or music. What would you like?"

    recommendation = random.choice(recommendations[category])
    response_text = f"I recommend {recommendation}."
    text_to_speech(response_text)

    return response_text
```

```python

def main():
    """Main function that listens to the user's voice and generates a response."""
    print("ðŸŽ™ Listening for a recommendation request...")
    # user_text = recognize_speech()
    print("Enter a recommendation request (e.g., 'I want a movie recommendation'): ")
    user_text = input().lower()
    if user_text:
        response = recommend(user_text)
        print("ðŸ¤– Chatbot:", response)
```

```python
if __name__ == "__main__":
    main()
```
