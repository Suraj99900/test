##### Step 5: Handle Outliers

```python
dMoviesData['budget'].info()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dMoviesData['budget'] = pd.to_numeric(dMoviesData['budget'], errors='coerce')
```

```python
# Checking the Outliers Between Budget revenu
Q1 = dMoviesData['budget'].quantile(0.25)
Q3 = dMoviesData['budget'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
```

```python
print("Q1",Q1)
print("Q3",Q3)
print("IQR",IQR)
print("lower_bound",lower_bound)
print("upper_bound",upper_bound)
print("shape Of data :- ", dMoviesData.shape)
```

```python
dMoviesData = dMoviesData[(dMoviesData['budget'] >= lower_bound)]
dMoviesData = dMoviesData[(dMoviesData['budget'] <= upper_bound)]
dMoviesData.shape
```
