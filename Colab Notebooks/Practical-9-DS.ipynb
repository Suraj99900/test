```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
oData = sns.load_dataset("titanic")
```

```python

oData.head()
```

```python
NewTitanic = oData.dropna(subset=['age', 'sex', 'survived'])
```

```python
NewTitanic.shape
```

```python
NewTitanic['survived'] = NewTitanic['survived'].map({0: 'Not Survived', 1: 'Survived'})
```

```python
NewTitanic.head()
```

```python
plt.figure(figsize=(10, 6))
plt.pie(NewTitanic['survived'].value_counts(), labels=NewTitanic['survived'].value_counts().index, autopct='%1.1f%%')
plt.title('Survival Distribution')
plt.show()
```

```python
# Plot boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(x='sex', y='age', hue='survived', data=NewTitanic)
plt.title('Age Distribution by Gender and Survival Status')
plt.xlabel('Sex')
plt.ylabel('Age')
plt.legend(title='Survival Status')
plt.grid(True)
plt.tight_layout()
plt.show()
```

Younger individuals had higher survival rates, especially noticeable among females. The median age of survived females is lower than those who did not survive.

Among males, the survival rate was generally lower, and there is less difference in median age between those who survived and those who didnt, although younger males still had a slightly better chance.

Outliers (dots outside the whiskers) show a wide range of ages among non-survivors, especially in males.

The interquartile range (IQR) for females who survived is relatively narrow and skewed towards younger ages.

This supports the historical context of the Titanic disaster where the policy "women and children first" might have influenced survival outcomes.
