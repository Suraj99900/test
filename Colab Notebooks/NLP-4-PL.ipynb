```python
import pandas as pd
import re

# Load the dataset
train_path = '/content/train.csv'
df = pd.read_csv(train_path)

# Function to clean text using regex
def clean_text(text):
    """Removes special characters, numbers, and extra spaces from the text."""
    text = re.sub(r'[^a-zA-Z\s]', '', text)  # Remove special characters and numbers
    text = re.sub(r'\s+', ' ', text).strip()  # Remove extra spaces
    return text

def search_toxic_words(text):
    """Searches for toxic words in a given text using real toxic words from the dataset."""
    toxic_samples = df[df[['toxic', 'severe_toxic', 'obscene', 'threat', 'insult', 'identity_hate']].sum(axis=1) > 0]
    toxic_comments = ' '.join(toxic_samples['comment_text'].dropna().astype(str))
    toxic_words = list(set(re.findall(r'\b\w{4,}\b', toxic_comments)))  # Extract words with 4+ letters
    found_words = [word for word in toxic_words if re.search(rf'\b{word}\b', text, re.IGNORECASE)]
    return found_words

# Example usage
test_text = "Why the edits made under my username Hardcore Metallica Fan were reverted? They weren't vandalisms, just closure on some GAs after I voted at New York Dolls FAC. And please don't remove the template from the talk page since I'm retired now.89.205.38.27"
cleaned_text = clean_text(test_text)
toxic_found = search_toxic_words(cleaned_text)
print("Cleaned Text:", cleaned_text)
print("Toxic Words Found:", toxic_found)

# Apply cleaning and toxic word search on dataset
df['cleaned_comment'] = df['comment_text'].apply(clean_text)
df['toxic_words'] = df['cleaned_comment'].apply(search_toxic_words)
display(df[['comment_text', 'cleaned_comment', 'toxic_words']].head())
```
