```python
import seaborn as sns
import pandas as pd
```

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
# Checking data types
print(iris.dtypes)
```

```python
import matplotlib.pyplot as plt
```

```python
iris.hist(figsize=(10, 8), bins=15, edgecolor='black')
plt.suptitle("Feature Distributions - Iris Dataset")
plt.tight_layout()
plt.show()
```

```python
iris.columns[:-1]
```

```python
plt.figure(figsize=(5,5))
sns.pairplot(iris, hue='species')
plt.suptitle("Pairwise Feature Distributions by Species", y=1.02)
plt.show()
```

This plot clearly shows that:

Setosa is linearly separable from the others.

Versicolor and Virginica overlap in petal dimensions, but can still be separated reasonably.

```python
plt.figure(figsize=(12, 8))
for i, column in enumerate(iris.columns[:-1], 1):
    plt.subplot(2, 2, i)
    sns.boxplot(x=iris[column])
    plt.title(f'Boxplot of {column}')
plt.tight_layout()
plt.show()
```

Observations:

Sepal Length: Fairly symmetric, few mild outliers.

Sepal Width: Slight left skew; a few clear outliers on the lower end.

Petal Length & Petal Width: Skewed distribution with distinct clusters (likely due to different species). Petal width shows a few potential outliers.
