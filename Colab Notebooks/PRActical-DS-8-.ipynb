```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
oData = sns.load_dataset('titanic')
```

```python
oData.head()
```

```python
oData.isnull().sum()
```

```python
oData.shape
```

```python
# Cleaning the data set
oData['deck'].unique()
```

```python
# Replace Nan with type code in deck
oData['deck'] = oData['deck'].fillna(oData['deck'].mode()[0])
```

```python
oData['age'] = oData['age'].fillna(oData['age'].mean());
oData['embarked'] = oData['embarked'].fillna(oData['embarked'].mode()[0]);
oData['embark_town'] = oData['embark_town'].fillna(oData['embark_town'].mode()[0]);
```

```python
oData.isnull().sum()
```

```python
oData.head()
```

```python
oData.info()
```

```python
# Addeing Label
import sklearn
from sklearn.preprocessing import LabelEncoder
```

```python
leSex = LabelEncoder()
leEmbarked = LabelEncoder()
leDeck = LabelEncoder()
leClass = LabelEncoder()
leEmbark_town = LabelEncoder()
leAlive = LabelEncoder()



```

```python
oData['sex'] = leSex.fit_transform(oData['sex'])
oData['embarked'] = leEmbarked.fit_transform(oData['embarked'])
oData['deck'] = leDeck.fit_transform(oData['deck'])
oData['class'] = leClass.fit_transform(oData['class'])
oData['embark_town'] = leEmbark_town.fit_transform(oData['embark_town'])
oData['alive'] = leAlive.fit_transform(oData['alive'])
```

```python
oData.info()
```

```python
oData.head()
```

```python
oData['adult_male'] = oData['adult_male'].astype(int)
oData['alone'] = oData['alone'].astype(int)
```

```python
oData.head()
```

```python
oNewData = oData.drop(['who'],axis=1)
```

```python
oTotalCorr = oNewData.corr()
```

```python
oTotalCorr
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(oTotalCorr,annot=True,cmap='coolwarm')
plt.show()
```

```python
sns.histplot(data=oNewData, x='fare', kde=True, bins=40, color='skyblue')
plt.title('Distribution of Fare')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.boxplot(x='class', y='fare', data=oNewData)
plt.title('Fare by Passenger Class')
plt.show()
```

```python
sns.boxplot(x='survived', y='fare', data=oNewData)
plt.title('Fare vs Survival Status')
plt.xticks([0, 1], ['Not Survived', 'Survived'])
plt.show()
```

```python
sns.violinplot(x='sex', y='fare', data=oNewData)
plt.title('Fare Distribution by Sex')
plt.show()
```

```python
sns.stripplot(x='class', y='fare', data=oNewData, jitter=True, hue='sex', dodge=True)
plt.title('Fares by Class and Sex')
plt.legend(loc='upper right')
plt.show()
```
