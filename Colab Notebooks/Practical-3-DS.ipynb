```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
```

```python
dfData = sns.load_dataset('titanic')
```

```python
dfData.head()
```

```python
dfData.shape
```

Compute Summary Statistics

```python
summary_stats = dfData.groupby("pclass")[["age", "survived"]].agg(["mean", "median", "min", "max", "std"])
summary_stats
```

Step 3: Convert Categorical Variable into Numeric Values

```python
pclass_numeric = dfData["pclass"].tolist()
print(pclass_numeric[:10])  # Display first 10 values
```

```python
# Basis analysis Process
dfData.isnull().sum()
```

```python
dfData['age'] = dfData['age'].fillna(dfData['age'].mean())
```

```python
dfData.shape
```

```python
dfData['deck'].unique()
```

```python
dfData['deck'].dropna(inplace=True)
```

```python
dfData.shape
```

```python
dfData['deck'].unique()
```

```python
dfData.dropna(subset=['deck'], inplace=True)
dfData.dropna(subset=['embarked'], inplace=True)
dfData.dropna(subset=['embark_town'], inplace=True)
```

```python
dfData
```

```python
dfData.info()
```

```python
dfData
```

```python
LabelEncoderData = LabelEncoder()
dfData['sex'] = LabelEncoderData.fit_transform(dfData['sex'])
dfData['embarked'] = LabelEncoderData.fit_transform(dfData['embarked'])
dfData['alive'] = LabelEncoderData.fit_transform(dfData['alive'])
dfData
```

```python
df_numerical_data = dfData.select_dtypes(include=[np.number])
df_numerical_data
```

Compute Correlation Matrix

```python
corr_matrix = df_numerical_data.corr()
corr_matrix
```

Step 4: Plot Heatmap

```python
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='PuOr',fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
GroupWithSurvivedAndAge = dfData.groupby('survived')['age'].agg(['mean', 'median', 'min', 'max', 'std'])
GroupWithSurvivedAndAge
```

From the Above table we can Say that.
#### 1) Average Age : 32 is for (Survived) or 39 is for (non Survived)
#### 2) Meadian Age : This is the similar case 38 for non survived and 31.5 for survived
#### 3) age range : The age who did not survived are 2-71 and how able to survived are 0.92 to 80

`This table suggests that younger passengers were more likely to survive the Titanic disaster, which aligns with the historical accounts that prioritized saving women and children. The mean and median ages for survivors are notably lower than those who did not survive.`

```python
sns.boxplot(x='survived', y='age', data=dfData)
plt.title('Age Distribution by Survival')
plt.xlabel('Survived (o = no , 1 = yes)')
plt.ylabel('Age')
plt.show()
```

```python
# Finding the Outliear in survived column
Q1 = dfData['survived'].quantile(0.25)
Q3 = dfData['survived'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
dfData[(dfData['survived'] < lower_bound) | (dfData['survived'] > upper_bound)]
```

From the Above code we ca see that there is now outlier in the survived column now check similar sinario for age

```python
# Finding the Outliear in age column
Q1 = dfData['age'].quantile(0.25)
Q3 = dfData['age'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
dfData[(dfData['age'] < lower_bound) | (dfData['age'] > upper_bound)]
```

```python
print("Min Age :- ", dfData['age'].min())
print("Max Age :- ", dfData['age'].max())
print("lower_bound :- ",lower_bound)
print("upper_bound :- ",upper_bound)
```

```python
# Filter data for non-survivors
non_survivors = dfData[dfData['survived'] == 0]
sns.histplot(non_survivors['age'], kde=True, color='blue', label='Non-Survivors')
plt.title('Age Distribution by Survival')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
# Filter data for survivors
survivors = dfData[dfData['survived'] == 1]

sns.histplot(survivors['age'], kde=True, color='green', label='Survivors')
plt.title('Age Distribution by Survival')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```
